## Bike Sharing dataset

This dataset has been taken from UCI Machine Learning Repository. The source can be found here: <a href='https://archive.ics.uci.edu/ml/datasets/bike+sharing+dataset'>Link</a>

**This dataset is from 2011-2012**

Context: Bike sharing systems are new generation of traditional bike rentals where whole process from membership, rental and return back has become automatic. Through these systems, user is able to easily rent a bike from a particular position and return back at another position. Currently, there are about over 500 bike-sharing programs around the world which is composed of over 500 thousands bicycles. Today, there exists great interest in these systems due to their important role in traffic, environmental and health issues.

Apart from interesting real world applications of bike sharing systems, the characteristics of data being generated by these systems make them attractive for the research. Opposed to other transport services such as bus or subway, the duration of travel, departure and arrival position is explicitly recorded in these systems. This feature turns bike sharing system into a virtual sensor network that can be used for sensing mobility in the city. Hence, it is expected that most of important events in the city could be detected via monitoring these data.



**Code Book**

- `instant`: record index
- `dteday` : date
- `season` : season (1:winter, 2:spring, 3:summer, 4:fall)
- `hr`: hour of day
- `holiday` : weather day is holiday or not (extracted from [Web Link])
- `workingday` : if day is neither weekend nor holiday is 1, otherwise is 0.
- `weathersit` :
  - 1: Clear, Few clouds, Partly cloudy, Partly cloudy
  - 2: Mist + Cloudy, Mist + Broken clouds, Mist + Few clouds, Mist
  - 3: Light Snow, Light Rain + Thunderstorm + Scattered clouds, Light Rain + Scattered clouds
  - 4: Heavy Rain + Ice Pallets + Thunderstorm + Mist, Snow + Fog
- `temp` : Normalized temperature in Celsius. The values are derived via (t-t_min)/(t_max-t_min), t_min=-8, t_max=+39 (only in hourly scale)
- `atemp`: Normalized feeling temperature in Celsius. The values are derived via (t-t_min)/(t_max-t_min), t_min=-16, t_max=+50 (only in hourly scale)
- `hum`: Normalized humidity. The values are divided to 100 (max)
- `windspeed`: Normalized wind speed. The values are divided to 67 (max)
- `cnt`: count of total rental bikes including both casual and registered

```python
#import packages
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Load the data
data = pd.read_csv('hour.csv')
```

```python
data.info()
```

```python
#Rename Columns
rename_map = {'instant' : 'id',
              'dteday' : 'timestamp',
              'season' : 'season',
              'hr' : 'hour',
              'holiday' : 'holiday',
              'workingday' : 'workingday',
              'weathersit' : 'weather',
              'temp' : 'temp',
              'atemp' : 'temp_feel',
              'hum' : 'humidity',
              'windspeed' : 'windspeed',
              'cnt' : 'demand'}
```

```python
data = data.rename(columns=rename_map)
```

```python
display(data.head())
```

```python
#Map Season and Weather to Categories
data['weather'] = data['weather'].map({1:'Clear', 2:'Misty + Cloudy', 3: 'Light Snow', 4: 'Heavy Rain + Ice'})
data['season'] = data['season'].map({1:'Winter', 2:'Spring', 3: 'Summer', 4: 'Fall'})
```

```python
#Convert Timestamp to Datetime Index
data = data.set_index('timestamp')
data.index = pd.to_datetime(data.index, format = '%m/%d/%Y')
```

```python
data.head()
```

```python
missing = pd.concat([data.isnull().sum(), 100 * data.isnull().mean()], axis = 1)
missing.columns = ['count', '%']
missing.sort_values(by = 'count', ascending = False)
```

```python
#Checking the correlation between temp and temp_feel
data.loc[:, ['temp', 'temp_feel']].corr()
```

```python
#Dropping the temp column
data.drop(columns='temp', inplace=True)
```

```python
data[['season', 'holiday', 'workingday', 'weather']] = data[['season', 'holiday', 'workingday', 'weather']].astype('object')
```

```python
def investigate_object(df):
    """
    This function investigates object columns in the dataset
    
    Parameters
    ----------
    df : Expects a pandas DataFrame
    
    Returns
    -------
    prints unique values
    """
    
    object_cols = df.select_dtypes('object').columns
    [print(df[col].value_counts(), '\n---------------------------------------') for col in df[object_cols]]
```

```python
investigate_object(data)
```

Note that **Heavy Rain + Ice** only has 3 values. This is a outlier and we should get rid of these rows altogether.

```python
#Drop the row with heavy rain / ice pellets
data = data[data.weather != 'Heavy Rain + Ice']
```

```python
#Plot Demand by Hour
%matplotlib inline
fig, ax = plt.subplots(2, 2, figsize=(20, 16))
plt.subplots_adjust(hspace=0.5);
#sns.lineplot(x = data.index.hour, y = 'demand' , data = data, hue = 'weather')
sns.lineplot(x = 'hour', y = 'demand' , data = data, hue = 'workingday', ax= ax[0][0])
ax[0][0].set_title('Demand of Bikes by Hour on Working Day')
sns.lineplot(x = 'hour', y = 'demand' , data = data, hue = 'holiday', ax= ax[0][1])
ax[0][1].set_title('Demand of Bikes by Hour on Holidays')
sns.lineplot(x = 'hour', y = 'demand' , data = data, hue = 'weather', ax= ax[1][0])
ax[1][0].set_title('Demand of Bikes by Hour on Weather')
sns.lineplot(x = 'hour', y = 'demand' , data = data, hue = 'season', ax= ax[1][1])
ax[1][1].set_title('Demand of Bikes by Hour on Season')
```

<strong><h3> Some interesting findings </strong></h3>Clearly, the trends show us that the maximum usage is between 8 a.m. and 7.00 p.m. 

1. Furthermore, you can see that the demand for bikes is much greater during peak commuting hours on Workdays and this makes sense because most people would be using the bike to commute to work. On Weekends however, you can clearly see that Demand increases gradually since many people would be relaxing in the morning and then step out for exercising. This would imply that people use bikes for exercising on Weekends rather than commuting to work. 

2. Demand based on Weather - The graphs make sense! When the weather is clear, you'd expect more people to use bikes than only light snowly days.

3. Season - One  finding is that demand for bikes is the lowest during Winter season. This makes sense because as the temperatures decrease, people tend to use the Bikes less. Demand is highest during the Summer and Spring time, and this makes sense

**Clearly, hour of the day matters significantly to predict demand and we should include this in the model**

```python
#Let's plot the histograms of numeric features to see the distribution
data.drop(columns = 'id').hist(figsize = (15, 15))
plt.subplots_adjust(hspace=0.5);
```

```python
#Running a heatmap of the correlations
plt.subplots(figsize = (12, 10))
sns.heatmap(data.drop(columns = 'id').corr())
```

Based on the correlations, `temp_feel` and `demand` are positively correlated. `humidity` on the other hand is negatively correlation. This makes sense because riders will not like to rides bikes when the humidity is high. Plus the chance of rainfall and sweat is high too.
